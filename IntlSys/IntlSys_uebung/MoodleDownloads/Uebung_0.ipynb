## Python
Hier lernen Sie erste Schritte in Python kennen. Zellen, die mit ```In [X]``` anfangen, enthalten Python-Code, der direkt über den ```Run```-Button ausgeführt werden können. Z.B. rechnet der ```**```-Operator den Wert der Potenz von $a^b$ aus.

````python
a = 2
b = 5
c = a**b
print(c)
````

Eine große Stärke von Python sind die unzähligen Pakete mit Programmbibliotheken, die sehr einfach verwendet werden können. Pakete werden mit ```import``` eingebunden. Das funktiuniert natürlich nur mit Paketen, die auch installiert wurden (durch die Admins). Die für die Übungen notwendigen Pakete sollten alle installiert sein. 

Für wissenschaftliches Rechnen kann z.B. das Paket [NumPy](https://numpy.org/) verwendet werden. Eingebunden wird es über: 


````python
import numpy as np
````

Damit können dann ab sofort die Funktionen der ```numpy```-Bibliothek über die "Abkürzung" ```np``` zugegriffen werden. Z.B. um 101 Werte mit identischem Abstand zwischen 0 und 10 zu erzeugen: 

````python
x = np.linspace(0, 10, 101)
print(x)
````

mit den Werten in x kann jezt weitergerechnet werden:

````python
y = 4 + 2 * np.sin(2 * x)
````

Und mit [matplotlib](https://matplotlib.org/) kann das Ganze dann geplottet werden:

````python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
````
