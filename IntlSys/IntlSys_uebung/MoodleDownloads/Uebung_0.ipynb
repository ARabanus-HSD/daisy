## Python
Hier lernen Sie erste Schritte in Python kennen. Zellen, die mit ```In [X]``` anfangen, enthalten Python-Code, der direkt Ã¼ber den ```Run```-Button ausgefÃ¼hrt werden kÃ¶nnen. Z.B. rechnet der ```**```-Operator den Wert der Potenz von $a^b$ aus.

Eine groÃŸe StÃ¤rke von Python sind die unzÃ¤hligen Pakete mit Programmbibliotheken, die sehr einfach verwendet werden kÃ¶nnen. Pakete werden mit ```import``` eingebunden. Das funktiuniert natÃ¼rlich nur mit Paketen, die auch installiert wurden (durch die Admins). Die fÃ¼r die Ãœbungen notwendigen Pakete sollten alle installiert sein. 

FÃ¼r wissenschaftliches Rechnen kann z.B. das Paket [NumPy](https://numpy.org/) verwendet werden. Eingebunden wird es Ã¼ber: 


````python
import numpy as np
````

Damit kÃ¶nnen dann ab sofort die Funktionen der ```numpy```-Bibliothek Ã¼ber die "AbkÃ¼rzung" ```np``` zugegriffen werden. Z.B. um 101 Werte mit identischem Abstand zwischen 0 und 10 zu erzeugen: 

````python
x = np.linspace(0, 10, 101)
print(x)
````

mit den Werten in x kann jezt weitergerechnet werden:

````python
y = 4 + 2 * np.sin(2 * x)
````

Und mit [matplotlib](https://matplotlib.org/) kann das Ganze dann geplottet werden:

````python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
````
