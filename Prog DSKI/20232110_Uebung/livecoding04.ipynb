```python
import math
import turtle

        
#------------------------------------------------------------------------
#------------------------------------------------------------------------

class GeometrischesObjekt:
    def __init__(self, starting_angle=0, position=(0, 0)) -> None:
        self.position = position
        self.starting_angle = starting_angle

    def move_to(self):
        turtle.penup()
        turtle.goto(self.position[0], self.position[1])
        turtle.pendown()
        turtle.setheading(self.starting_angle)

#------------------------------------------------------------------------
#------------------------------------------------------------------------
        
class RegularPolygon(GeometrischesObjekt):
    def __init__(self, edge_length:float, corner_count:float, starting_angle=0, position=(0, 0)):
        self.edge_length = edge_length
        self.corner_count = corner_count
        self.position = position
        self.starting_angle = starting_angle
        
        # self.geometrisches_objekt = GeometrischesObjekt(self.starting_angle, self.position)
        pass
    
    
    def draw(self):
        for i in range(self.corner_count):
            turtle.forward(self.edge_length)
            angle = 360/self.corner_count
            turtle.left(angle)
        pass
    
    def area(self):
        """
        draw a triangle from the "center" to the middle of each edge
        height of triangle:
            h = (edge/2) / (tan(alpha/2))
        """
        angle = 360/self.corner_count
        h = (self.edge_length/2) / math.tan(math.radians(angle/2))
        print(f"distance from center to middle of edge is {h}")
        area = self.corner_count * (self.edge_length * h)/2
        print(f"area of this regular polygon is {area}")
        pass
    
    

        
#------------------------------------------------------------------------
#------------------------------------------------------------------------

class Square(RegularPolygon):
    def __init__(self, edge_length, starting_angle:float, position=(0, 0)):
        self.edge_length = edge_length
        self.corner_count = 4
        self.position = position
        self.starting_angle = starting_angle

        self.regular_polygon_instance = RegularPolygon(self.edge_length, self.corner_count)
        self.geometrisches_objekt = GeometrischesObjekt(self.starting_angle, self.position)
        pass
    
    
 
    
    
        
#------------------------------------------------------------------------
#------------------------------------------------------------------------

class Rectangle(GeometrischesObjekt):
    def __init__(self, edge_length_1, edge_length_2, starting_angle=0, position=(0, 0)):
        self.edge_length_1 = edge_length_1
        self.edge_length_2 = edge_length_2
        self.corner_count = 4
        self.position = position
        self.starting_angle = starting_angle
        
        self.geometrisches_objekt = GeometrischesObjekt(self.starting_angle, self.position)
        pass
    
    
    def draw(self):
        for i in range(2):
            turtle.forward(self.edge_length_1)
            angle = 90
            turtle.left(angle)
            turtle.forward(self.edge_length_2)
            turtle.left(angle)
        pass
    
    def area(self):
        area = self.edge_length_1 * self.edge_length_2
        print(f"area of rectangle is {self.edge_length_1} * {self.edge_length_2} = {area}")
        pass
        
#------------------------------------------------------------------------
#------------------------------------------------------------------------

class Triangle(RegularPolygon):
    
    def __init__(self, edge_length, starting_angle:float, position=(0,0)):
        self.edge_length = edge_length
        self.corner_count = 3
        self.position = position
        self.starting_angle = starting_angle
        
        self.regular_polygon_instance = RegularPolygon(self.edge_length, self.corner_count)
        self.geometrisches_objekt = GeometrischesObjekt(self.starting_angle, self.position)
        pass
  
    
#------------------------------------------------------------------------
#------------------------------------------------------------------------




def main():
    # ----- IGNORE THIS PART ---------------------------------
    wn = turtle.Screen()
    rootwindow = wn.getcanvas().winfo_toplevel()
    rootwindow.call('wm', 'attributes', '.', '-topmost', '1')
    rootwindow.call('wm', 'attributes', '.', '-topmost', '0')
    # ----- IGNORE THIS PART ---------------------------------
    
    
    
    # AUFGABE 2.2 CODING COMES HERE
    
    
    shape_list = [
        RegularPolygon(100, 6, 45, (100, 100)),
        Square(100, 456, (-100, -100)),
        Triangle(100, 12, (0, 1)),
        Rectangle(100, 50, 0, (-100, 100))]
        
    for shape in shape_list:
        shape.move_to()
        shape.draw()
        shape.area()

    
    # ----- IGNORE THIS PART ---------------------------------
    wn.mainloop()
    turtle.done()
    # ----- IGNORE THIS PART ---------------------------------

main()
```
